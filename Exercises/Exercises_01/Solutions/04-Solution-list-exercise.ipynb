```python
#1.Dice rolls.

import random as rnd
dice_rolls = 10
ls_dice =[]

for i in range(dice_rolls):
    dice = rnd.randint(1,6)
    ls_dice.append(dice)
ls_dice_sort = sorted(ls_dice)
ls_dice_sort_reverse = sorted(ls_dice, reverse=True)

print(f"Dice rolls list: {ls_dice}.")
print(f"Dice rolls list in ascending order: {ls_dice_sort}")
print(f"Dice rolls list in descending order: {ls_dice_sort_reverse}")
print(f"Maximum value: {max(ls_dice)}")
print(f"Minimum value: {min(ls_dice)}")
```

```python
#5a. Dice rolls convergence.

import random as rnd
dice_rolls = 100
count = 0

for i in range(dice_rolls):
    dice = rnd.randint(1,6)
    if dice == 6:
        count+=1
print(f"The number of outcome six in 100 dice rolls is: {count}.")
```

```python
#5b. Dice rolls convergence.

import random as rnd
dice_rolls = 10
count = 0
count_six = []
prob_six = []

while dice_rolls <= 1000000:
    for i in range(dice_rolls):
        dice = rnd.randint(1,6)
        if dice == 6:
            count+=1
    prob = count/dice_rolls
    count_six.append(count)
    prob_six.append(prob)
    dice_rolls*=10
    
print(f"Count number of six: {count_six}")
print(f"P(six) = {prob_six}")
```

```python
#5c. Dice rolls convergence.

import random as rnd
import matplotlib.pyplot as plt
import matplotlib.ticker as tiker

rnd.seed(1)
dice_rolls = 10
count = 0
count_six = []
prob_six = []

while dice_rolls <= 1000000:
    for i in range(dice_rolls):
        dice = rnd.randint(1,6)
        if dice == 6:
            count+=1
    prob = count/dice_rolls
    count_six.append(count)
    prob_six.append(prob)
    dice_rolls*=10
    
print(f"Count number of six: {count_six}")
print(f"P(six) = {prob_six}")

plt.plot(prob_six, '-*')
plt.title("Probability of six for different number of rolls")
#plt.xticks([0,1,2,3,4,5], dice_rolls)
plt.xlabel("Number of dice rolls")
plt.ylabel("Probability")

plt.show()
```
